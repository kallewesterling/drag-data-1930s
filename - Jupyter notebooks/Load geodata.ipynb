```python
import datetime
from pandas import to_datetime
from assistance.data import get_clean_network_data
from assistance.constants import urls
from assistance.geo import get_google_geodata, get_nomatim_geodata
```

```python
df = get_clean_network_data(verbose=False)
```

```python
df['Year'] = df.apply(lambda row: to_datetime(row.Date).year, axis=1)
df['Lat'] = df.apply(lambda row: get_google_geodata(row.City).get('geometry', {}).get('location', {}).get('lat', ''), axis=1)
df['Lng'] = df.apply(lambda row: get_google_geodata(row.City).get('geometry', {}).get('location', {}).get('lng', ''), axis=1)
```

```python
geo_df = df[(df['Lat'] != '') & (df['Lng'] != '') & (df['Year'] >= 1930) & (df['Year'] <= 1940)]
```

```python
for groups, rows in geo_df.sort_values('Date').groupby(['Performer', 'Year', 'Lat', 'Lng']):
    performer, year, lat, lng = groups
    print(performer, year, lat, lng)
```
